```python
import pandas as pd
```

```python
df = pd.read_excel("input/Riskassessment_AU LL.xlsx")
df
```

```python
df.columns
```

```python
small = df.dropna(subset= "Unnamed: 3")[[ 'Riskassesment  for infectious disease spreading in GBN',"Unnamed: 3","Unnamed: 4","Unnamed: 5"]]
small.columns = ["topic","done","todo","comments"]
small = small.dropna(subset="topic").reset_index(drop=True)
small
```

```python

for ix, row in small.iterrows():
    if ix < 23:
        small.loc[ix,"category"] = "Technology centered disease control"
        if ix <= 10:
            small.loc[ix,"subcategory"] = "Air quality"
        elif ix < 17:
            small.loc[ix,"subcategory"] = "Water quality"
        else:
            small.loc[ix,"subcategory"] = "Sanitizing"
    elif ix < 44:
        small.loc[ix,"category"] = "Human centred disease control"
        if ix < 29:
            small.loc[ix,"subcategory"] = "User journey"
        elif ix <= 35:
            small.loc[ix,"subcategory"] = "User behavior"
        else:
            small.loc[ix,"subcategory"] = "Mental health"
    elif ix < 55 :
        small.loc[ix,"category"] = "Continuity"
        if ix <= 47:
            small.loc[ix,"subcategory"] = "Propagation monitoring"
        elif ix <= 52:
            small.loc[ix,"subcategory"] = "Continuity"
        else:
            small.loc[ix,"subcategory"] = "Information management"
```

```python
small.subcategory.unique()
```

```python
small.loc[small.subcategory == 'Air quality',"important"] = small.loc[3,"done"]
small.loc[small.subcategory == 'Water quality',"important"] = small.loc[11,"done"]
small.loc[small.subcategory == 'Sanitizing',"important"] = small.loc[17,"done"]

small.loc[small.subcategory == 'User journey',"important"] = small.loc[23,"done"]
small.loc[small.subcategory == 'User behavior',"important"] = small.loc[29,"done"]
small.loc[small.subcategory ==  'Mental health',"important"] = small.loc[36,"done"]

small.loc[small.subcategory == 'Air quality',"assessed"] = small.loc[10,"done"]
small.loc[small.subcategory == 'Water quality',"assessed"] = small.loc[16,"done"]
small.loc[small.subcategory == 'Sanitizing',"assessed"] = small.loc[22,"done"]

small.loc[small.subcategory == 'User journey',"assessed"] = small.loc[28,"done"]
small.loc[small.subcategory == 'User behavior',"assessed"] = small.loc[35,"done"]
small.loc[small.subcategory ==  'Mental health',"assessed"] = small.loc[43,"done"]

if 0:
    small.loc[small.subcategory == 'Propagation monitoring',"important"] = "done"
    small.loc[small.subcategory == 'Continuity',"important"] = small.loc[3,"done"]
    small.loc[small.subcategory == 'Information management',"important"] = small.loc[3,"done"]
```

```python
cleared = small[4:][["category","subcategory","important","assessed","topic","done","todo","comments"]].dropna(subset="todo")


cleared.loc[cleared.done == "No relevant measures in place","done_score"] = 0
cleared.loc[cleared.todo == "No relevant measures in place","todo_score"] = 0

cleared.loc[cleared.done == "In minor parts of the evaluated GBN","done_score"] = 1
cleared.loc[cleared.todo == "In minor parts of the evaluated GBN","todo_score"] = 1

cleared.loc[cleared.done == "In some parts of the evaluated GBN","done_score"] = 2
cleared.loc[cleared.todo == "In some parts of the evaluated GBN","todo_score"] = 2

cleared.loc[cleared.done == "In large parts of the evaluated GBN","done_score"] = 3
cleared.loc[cleared.todo == "In large parts of the evaluated GBN","todo_score"] = 3

cleared.loc[cleared.done == "Thorughout the evaluated GBN","done_score"] = 4
cleared.loc[cleared.todo == "Thorughout the evaluated GBN","todo_score"] = 4

cleared.loc[cleared.done == "No relevant measures in place","done_score"] = 0
cleared.loc[cleared.todo == "No relevant measures planned","todo_score"] = 0

cleared.loc[cleared.done == "Measures in place","done_score"] = 4
cleared.loc[cleared.todo == "Measures planned" ,"todo_score"] = 4

cleared["evolution"] = cleared.todo_score - cleared.done_score
cleared
```

```python
cleared.to_excel("analysis.xlsx")
```
