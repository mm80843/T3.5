```python
evalRisks = [
        {
        "name": "eval_risks",
        "description": "You are a world-wide risk expert examining at how to mitigate contagious diseases propagation in the built environment, at all scales. Out of the python list given to you, sort what items are clearly the name of a risk, as classify as 'non-risk' something that is not a risk (example impacts or single word names).",
        "parameters": {
            "type": "object",
            "properties": {
                "risks": {
                    "type": 'array',
                    "description": "The list of actual risks",
                    "items": {
                                "type": 'string', 
                                "description": 'item that is a risk'
                            }
                        
                    },

                "non-risks": {
                    "type": 'array',
                    "description": "The list of what are not risks",
                    "items": {
                                "type": 'string', 
                                "description": 'item that is not a risk'
                            }
                        
                    }
                },
            },
            "required": ["risks","non-risks"],
        }
    
]
```

```python
def ask(CONTEXT,QUESTION,prefix,functions,overwrite=False,MODEL="gpt-3.5-turbo-1106"):
    REQ = {
        "context": CONTEXT,
        "question": QUESTION,
        "model": MODEL,
        "functions": functions,
        "token": TOK3N,
        "overwrite": overwrite,
        "source": "local-evalrisksT3.5"
    }

    H = h.hashme(CONTEXT+QUESTION)
    cached = "cache/"+prefix+"_"+H+".json"
    if not os.path.isfile(cached) or overwrite:
        x = requests.post(URL+"fct/", json = REQ)
        answer = json.loads(x.text)["messages"][-1]
        h.svt(cached,answer)
    else:
        answer = h.ldt(cached)
    
    return answer
```

```python
def askQ(CONTEXT,QUESTION,prefix,overwrite=False,MODEL="gpt-3.5-turbo-1106"):
    REQ = {
        "context": CONTEXT,
        "question": QUESTION,
        "model": MODEL,
        "token": TOK3N,
        "overwrite": overwrite,
        "source": "local-evalrisksT3.5"
    }

    H = h.hashme(CONTEXT+QUESTION)
    cached = "cache/"+prefix+"_"+H+".json"
    if not os.path.isfile(cached) or overwrite:
        x = requests.post(URL+"ask/", json = REQ)
        answer = json.loads(x.text)["messages"][-1]
        h.svt(cached,answer)
    else:
        answer = h.ldt(cached)
    
    return answer
```

```python
NB = len(RISKS)//200
cR = []
cNR = []
for k in range(NB):
    try:
        Rs =  "* "+"\n* ".join(RISKS[k*200:(k+1)*200+3]) 
        answer = ask("Out the following list, identify which items are clearly risk names,  of specific risks, relevant for a city to mitigate: \n\n",Rs,prefix="risk_sort",functions=evalRisks,overwrite=True)
        if "risks" in json.loads(answer['function_call']["arguments"]).keys():
            [ cR.append(x) for x in json.loads(answer['function_call']["arguments"])["risks"]]
        if "non-risks" in json.loads(answer['function_call']["arguments"]).keys():
            [cNR.append(y) for y in json.loads(answer['function_call']["arguments"])["non-risks"]]
    except:
        pass
```

```python
CONTEXT = """You are given a list of risks. Classify them using 10 groups of risks you define. Answer only as a bullet point list structure, similar to:
## Name of risk group
- Risk A
- Risk C 
## Name of risk group
- Risk B
etc..
"""
NChunks = 1000

NB = len(RISKS)//NChunks
CATS = []
for k in range(NB):
    QUESTION =  "The risks are:\n\n* "+"\n* ".join(RISKS[(k*NChunks):(k+1)*NChunks]) 
    REQ = {
        "context": CONTEXT,
        "question": QUESTION,
        "token": TOK3N,
        "overwrite": True,
        "source": "local-evalrisksT3.5"
    }

    x = requests.post(URL+"ask/", json = REQ)
    CATS.append(json.loads(x.text)["answer"])
```
