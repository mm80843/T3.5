```python
for ix, row in df.iterrows():
    newBP =  u.addItem("Blueprint", row["Blueprint title"], dIDct, onto )
    newBP.has_BPDescription.append(row["Idea description"])
    newBP.has_BPTitle.append(row["Blueprint title"])
    newBP.has_BPReason.append(row["Why?"])
    newBP.has_BPContext.append(row["Contextual situations"])
    newBP.has_BPProblem.append(row["Problems to solve"])
    newBP.has_BPValueProposition.append(row["Value proposition"])
    newBP.has_BPCapabilities.append(row["Capabilities"])
    newBP.has_BPRestriction.append(row["Restrictions and limitations"])    
    newBP.has_BPDemonstrated.append(row["Demonstrated"])

    pt = [x.strip().title() for x in row["Permanent"].split("&")]
    for pti in pt:
        PT = u.addItem("BP_Permanent", pti, dIDct, onto )
        newBP.has_BPPermanent.append(PT)   

    TRstring = row["Transmission"].replace("All","Airborne & Respiratory & Direct Contact")
    TRstring = [x.strip().title() for x in TRstring.split("&")]
    for tri in TRstring:   
        TR = u.addItem("BP_Transmission", tri, dIDct, onto )
        newBP.has_BPTransmission.append(TR)

    #Building', 'Both', 'GBN '
    TRstring = row["Scale"].replace("Both","GBN & Building")
    TRstring = [x.strip().title() for x in TRstring.split("&")]
    for tri in TRstring: 
        SC = u.addItem("BP_Scale", tri, dIDct, onto )
        newBP.has_BPScale.append(TR)

    TRstring = row["Phase"].replace("All","Dismantling & Operational & Design & Construction")
    TRstring = TRstring.replace(" and "," & ")
    TRstring = [x.strip().title() for x in TRstring.split("&")]
    for tri in TRstring: 
        PH = u.addItem("BP_Phase", tri, dIDct, onto )
        newBP.has_BPPhase.append(PH)

    TRstring = [x.strip().title() for x in row["Technical or behavioural"].split("&")]
    for tri in TRstring:     
        TB = u.addItem("BP_Type", tri, dIDct, onto )
        newBP.has_BPType.append(PT) 
    

    STV = [x.split("ID")[-1].strip() for x in row["People who suffer from"].split("\n") if (len(x) and "ID" in x)]
    for stv in STV:
        for k in dIDct["Stakeholder"].copy():
            if dIDct["Stakeholder"][k] == int(stv): #k pour le nom
                st = u.addItem("Stakeholder", k, dIDct, onto )
                newBP.has_BlueprintSubject.append(st)
    STV = [x.split("ID")[-1].strip() for x in row["People who can implement"].split("\n") if (len(x) and "ID" in x)]
    for stv in STV:
        for k in dIDct["Stakeholder"].copy():
            if dIDct["Stakeholder"][k] == int(stv): #k pour le nom
                st = u.addItem("Stakeholder", k, dIDct, onto )
                newBP.has_BlueprintOwner.append(st)

    STV = [x.split("ID")[-1].strip() for x in row["Technological enablers"].split("\n") if (len(x) and "ID" in x)]
    for stv in STV:
        for k in dIDct["BP_Enabler"].copy():
            if dIDct["BP_Enabler"][k] == int(stv): #k pour le nom
                #print(stv,k)
                st = u.addItem("BP_Enabler", k, dIDct, onto )
                newBP.has_BPEnabler.append(st)

    STV = [x.split("ID")[-1].strip() for x in row["Associated interventions"].split("\n") if (len(x) and "ID" in x)]
    for stv in STV:
        for k in dIDct["BP_Intervention"].copy():
            if dIDct["BP_Intervention"][k] == int(stv): #k pour le nom
                #print(stv,k)
                st = u.addItem("BP_Intervention", k, dIDct, onto )
                newBP.has_BPIntervention.append(st)


    print("-->",row["Blueprint title"])
```
